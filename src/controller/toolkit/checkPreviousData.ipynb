## AngleList

```python
s2e_path = '/gpfs/exfel/data/user/grotec/2NIP_S2E/'
diff_30fs_Com = os.path.join(s2e_path,'sim_5kev_30fs_35_2NIP_EMC_Compton/diffr')
diff_30fs_noCom = os.path.join(s2e_path,'sim_5kev_30fs_35_2NIP_EMC_noCompton/diffr')
diff_3fs_noCom = os.path.join(s2e_path,'/gpfs/exfel/data/user/grotec/2NIP_S2E/5keV_3fs_2NIP/diffr/diffr_noCompton')
diff_3fs_Com = os.path.join(s2e_path,'/gpfs/exfel/data/user/grotec/2NIP_S2E/5keV_3fs_2NIP/diffr/diffr_Compton')

data_list = [diff_30fs_Com,
             diff_30fs_noCom,
             diff_3fs_Com,
             diff_3fs_noCom]

# Scan Quaternion
for i in range(1,10):
    for diff_data in data_list:
        toolkit.listOrientation(diff_data,i)
    print ()
```

## Plot Data

```python
noCom = toolkit.getData(diff_3fs_noCom,1)
Com = toolkit.getData(diff_3fs_Com,1)
fig, ax, im = toolkit.plotLog(noCom,vmin=1e-5,vmax=1e1)
fig.set_size_inches(10,10)
fig, ax, im = toolkit.plotLog(Com,vmin=1e-5,vmax=1e1)
fig.set_size_inches(10,10)
```
